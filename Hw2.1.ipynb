```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_excel("fertility_rate.xlsx")
#df = table.copy()
print(df.head())
```

```python
print(f'Verses: {df.shape[0]}; Columns: {df.shape[1]}')
```

```python
import numpy as np
df = df.replace('_', np.nan)
print(df.head())
```

```python
print(df.dtypes)
```

```python
df[2019] = df[2019].astype(float)
df[2021] = df[2021].astype(float)
print(df.dtypes)
```

```python
print(df.isnull().sum() / len(df))
```

```python
print(df)
df = df.drop([26])
print(df)
```

```python
print(df)
numeric_columns = df.select_dtypes(include='number')
filled = numeric_columns.fillna(numeric_columns.mean().round(2))
df = pd.concat([df['regions'], filled], axis=1)
print(df)
```

```python
mean = df[2019].sum() / len(df)
df_2019 = df[['regions', 2019]]
print(df_2019[df_2019[2019] > mean])
```

```python
df_2015 = df[['regions', 2015]]
print(df_2015.loc[df[2015].idxmax()])
```

```python
import matplotlib.pyplot as plt

plt.bar(df['regions'], df[2019])
plt.xticks(rotation=90)
plt.title('Fertility rate in 2019')
plt.xlabel('Region')
plt.ylabel('Value')
plt.show()
```

```python
melted = df.melt(id_vars=['regions'], var_name='Year', value_name='avg_value')
print(melted)
grouped = melted.groupby('Year').agg(avg=('avg_value', 'mean'))
print(grouped)
plt.plot(grouped, marker='o')
plt.title('Fertility rate in Ukraine')
plt.xlabel('Year')
plt.ylabel('Average fertility value')
plt.grid(True)
plt.show()
```

```python
df['average'] = df.drop(columns='regions').mean(axis=1)
max_avg = df[df['average'] == df['average'].max()]
min_avg = df[df['average'] == df['average'].min()]

new = pd.concat([min_avg, max_avg])
new = new.drop(['average'], axis=1)
new.set_index(['regions'], inplace=True)
print(new)

plt.style.use('dark_background')

for region in new.index:
    plt.plot(new.columns, new.loc[region], marker='o', label=region)

plt.title('Comparision in fertility rate between highest and lowest averages')
plt.xlabel('Year')
plt.ylabel('Fertility rate')
plt.legend(title='region')
plt.grid(True)
plt.show()
```
